## An Extension To Imputation

```python
sorted_nan
```

```python
drop_test_col = [tup[0] for tup in sorted_nan[:6]]
drop_test_col
```

```python
imp_test_col = [tup[0] for tup in sorted_nan[6:]]
imp_test_col
```

```python
dropped_test = test.drop(drop_test_col, axis = 1)
dropped_train = train.drop(drop_test_col, axis = 1)
```

```python
# make copy to avoid changing original data (when Imputing)
new_test = dropped_test.copy()

for col in imp_test_col:
    new_test[col + '_was_missing'] = new_test[col].isnull()

# Imputation
my_imputer = Imputer()
new_test = pd.DataFrame(my_imputer.fit_transform(new_test))
new_test.columns = dropped_test.columns
```

```python
# make copy to avoid changing original data (when Imputing)
new_train = dropped_train.copy()

for col in imp_train_col:
    new_train[col + '_was_missing'] = new_train[col].isnull()

# Imputation
my_imputer = SimpleImputer()
new_train = pd.DataFrame(my_imputer.fit_transform(new_train))
new_train.columns = dropped_train.columns
```

# Modelling

```python
reduced_test = test.drop(test_li, axis = 1)
reduced_test.head()
```

```python
reduced_train = train.drop(test_li, axis = 1)
reduced_train = reduced_train.drop('SalePrice', axis = 1)
reduced_train.head()
```

```python
# converting the categorical data of test
test_predictors = pd.get_dummies(reduced_test)
test_predictors.head()
```

```python
# converting the categorical data of train
train_predictors = pd.get_dummies(reduced_train)
train_predictors.head()
```

```python
# to ensure the test data is encoded in the same manner
final_train_X, final_test_X = train_predictors.align(test_predictors,
                                                 join = 'left', axis = 1)
```

```python
# in order to isolate the price
final_train_y = train.SalePrice
```

```python
len(final_train_X)
```

```python
my_model = XGBRegressor(n_estimators = 1000, learning_rate=0.05)
my_model.fit(final_train_X, final_train_y)#, early_stopping_rounds = 5)
             #, eval_set = [(test_X, test_y)], verbose = False)
```

```python
a = my_model.predict(final_test_X)

#for price in a:
#    print(price)
list(a)
```
